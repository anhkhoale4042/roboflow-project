```python
from getpass import getpass
ROBOFLOW_API_KEY = getpass("Paste your Roboflow API key: ")

from roboflow import Roboflow
from pathlib import Path

rf = Roboflow(api_key=ROBOFLOW_API_KEY)
project = rf.workspace("count-cake").project("count-cake-4x4-grjfb")
dataset = project.version(3).download("yolov5", location="datasets/count-cake-4x4")

print("Dataset downloaded successfully!")
```

```python
from pathlib import Path

base = Path("datasets/count-cake-4x4")

for split in ["train","valid","test"]:
    imgs = list(base.rglob(f"*{split}*/**/*.jpg")) + list(base.rglob(f"*{split}*/**/*.png"))
    print(f"{split}: {len(imgs)} images")
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import random

sample_imgs = random.sample(list(base.rglob("train/**/*.jpg")), 3)
for img_path in sample_imgs:
    img = Image.open(img_path)
    plt.imshow(img)
    plt.title(img_path.name)
    plt.show()
```
